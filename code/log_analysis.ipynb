```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
log_data = pd.read_csv('logData.csv')
log_data[log_data['callerId']==2228796].head(20)
```

```python
log_group = log_data.groupby(by='callerId')
content_played = []
for item in log_data['logInfo']:
    if 'Content Played' in item:
        content_played.append(1)
    else:
        content_played.append(0)
log_data['content_played'] = content_played
```

```python
sns.histplot(data=log_data, x="langId", hue="content_played",multiple="dodge",discrete=True)
plt.xticks([0,1,2,3,4,5])
```
